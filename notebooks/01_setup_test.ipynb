```python
# Test connexion Albert API (optionnel)
import httpx
import os

albert_key = os.getenv("ALBERT_API_KEY")
albert_url = os.getenv("ALBERT_API_URL", "https://albert.api.etalab.gouv.fr/v1")

if albert_key and albert_key != "votre_cle_api_albert":
    try:
        response = httpx.get(
            f"{albert_url}/models",
            headers={"Authorization": f"Bearer {albert_key}"},
            timeout=10
        )
        if response.status_code == 200:
            models = response.json().get("data", [])
            print(f"✅ Connexion Albert API OK - {len(models)} modèles disponibles")
        else:
            print(f"⚠️  Albert API répond mais status {response.status_code}")
    except Exception as e:
        print(f"❌ Erreur connexion Albert API: {e}")
else:
    print("⏭️  Test Albert API ignoré (pas de clé configurée)")
```
